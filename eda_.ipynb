```python
plt.figure(figsize=(20,16))
corr = df.corr()
sb.heatmap(corr, annot=True)
plt.show()
plt.savefig('confusion_matrix.png')
```

```python
plt.figure(figsize=(20,16))
df.corr()['output'].sort_values().plot.barh()
plt.show()
plt.savefig('corgraph.png')
```
